```python
#Creamos un dataframe nuevo dejando solo las columnas que nos interesan.
df_preguntas=df.drop(columns=['case_number', 'type', 'area', 'location','activity', 'name', 'time', 'href', 'siglo','injury','date'])
df_preguntas.head()
```

2. ¬øHay valores duplicados en nuestro dataframe? En caso de que los haya, elimin√°ndlos.

```python
#Obtenemos las filas que estan dupicadas en nuestro dataframe.
df_preguntas[df_preguntas.duplicated()==True]
```

```python
#Observamos cuantas filas tenemos antes de eliminar los duplicados
print(f'Nuestro DataFrame antes de eliminar duplicados tiene {df_preguntas.shape[0]} filas y {df_preguntas.shape[1]} columnas')
```

```python
df_preguntas.drop_duplicates(inplace=True)
df_preguntas.head()
```

```python
#Observamos cuantas filas tenemos despues de eliminar los duplicados.
print(f'Nuestro DataFrame despues de eliminar duplicados tiene {df_preguntas.shape[0]} filas y {df_preguntas.shape[1]} columnas')
```

3. Como hemos visto, algunas columnas no tienen el tipo de datos que deber√≠an. Cambiad el tipo de dato para la columna de year.

```python
#Comprobamos el tipo de dato que tiene.
df_preguntas['year'].dtypes
```

```python
df_preguntas['year']=df_preguntas['year'].astype('int64', errors='ignore')
df_preguntas.head()
```

```python
#Comrpbamos que se ha cambiado el tipo de dato.
df_preguntas['year'].dtypes
```

4. En la columna de country poner todos los valores en min√∫sculas.
üí° Pista Tendr√©is que usar una funci√≥n o una lambda üí°

```python
#Creamos una funcion para pasar a minusculas la columna pais.
def pais_minusculas(columna):
    return str(columna).lower()
```

```python
#Aplicamos la funcion para cambiar a minusculas la columna country.
df_preguntas['country']=(df_preguntas['country'].apply(pais_minusculas))
df_preguntas.head()
```

```python
#Observamos que los valores unicos de la columna country estan todos en minusculas.
df_preguntas['month'].unique()
```
