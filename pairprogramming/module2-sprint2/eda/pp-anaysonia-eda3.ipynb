```python
#creamos el dataframe con los datos del csv obtenido en el pair de limpieza II.
df=pd.read_csv('files/attacks_limp2.csv',index_col=0)
df.head(3)
```

```python
#Creamos un countplot para ver el numero de ataques en cada pais.
sns.countplot(x='country',data=df)
plt.tick_params(labelrotation=45)
plt.title('Numero de ataques por pais')
plt.show();
```

```python
#Como observamos que no se puede ver claramente a que pais corresponde cada barra, y hay varios paises
#  en los que no podemos ver claro sus datos, nos quedamos con los 10 paises con m치s ataques.
df_mas_ataques=df['country'].value_counts()
df_mas_ataques.head(10)
```

```python
#Obtenemos la lista de los 10 paises con mas taques registrados.
df_10=df_mas_ataques.iloc[:10]
df_10.index[0:]
```

```python
#Filtramos nuestro DataFrame original seleccionando solo aquellas filas que correspondan los 10 paises con m치s ataques.
df_10_primeros=df[(df['country']=='usa')|(df['country']=='australia')|(df['country']=='south africa')|(df['country']=='bahamas')|(df['country']=='new zealand')|(df['country']=='reunion')|(df['country']=='brazil')|(df['country']=='papua new guinea')|(df['country']=='mexico')|(df['country']=='italy')]
df_10_primeros.head()
```

```python
#Comprobamos que se han seleccionado solo las filas de esos 10 paises.
df_10_primeros['country'].unique()
```

```python
#Creamos un countplot para ver el numero de ataques en esos 10 paises.
sns.countplot(x='country',data=df_10_primeros)
plt.tick_params(labelrotation=45)
plt.title('10 paises con mayor numero de ataques')
plt.show();
```

- Usando el DataFrame de los 10 paises con m치s ataques, 쯛ay diferencias entre sexos en esos paises?

```python
#Creamos de nuevo el mismo countplot para ver el numero de ataques en esos 10 paises, pero a침adimos el
#  parametro hue para ver la diferencia por sexos.
sns.countplot(x='country',data=df_10_primeros,hue='gender')
plt.tick_params(labelrotation=45)
plt.title('10 paises con mayor numero de ataques segun genero')
plt.show();
```

- 쯈u칠 edad recibe mayor n칰mero de ataques? 游눠 Pista 游눠 Para que esta gr치fica nos salga legible tendremos que crear antes grupos de edad. Para hacer esto de una forma sencilla tenemos el m칠todo pd.cut() que aprendimos en la lecci칩n de Limpieza II.

```python
#Creamos una columna con 5 rangos de edad.
df['rangos_edad']=pd.cut(df['edad'],bins=5)
df.head()
```

```python
#Creamos un countplot con los rangos de edad que hemos generado en la columna 'rangos_edad'.
sns.countplot(x='rangos_edad',data=df)
plt.tick_params(labelrotation=45)
plt.title('Numero de ataques por edades')
plt.show();
```

- Ataques entre sexos, 쯛ay diferencia?

```python
sns.countplot(x='gender',data=df)
plt.tick_params(labelrotation=45)
plt.title('Numero de ataques segun genero')
plt.show();
```

- 쮺uando ocurren m치s ataques? 쯆s sale algo raro en la gr치fica? Intentad limpiar un poco los datos para que os salga mejor.

```python
sns.countplot(x='month',data=df,order=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tick_params(labelrotation=45)
plt.title('Distribucion de los ataques a lo largo del a침o')
plt.show();
```

`Vemos que ocurren m치s ataques en los meses de verano (segun hemisferio norte), pero tambien hay un aumento de casos en los meses que corresponden al verano en el hemisferio sur, es decir, el invierno en el hemisferio norte.`

```python
sns.countplot(x='year',data=df)
plt.tick_params(labelrotation=45)
plt.title('Historico de ataques a nivel mundial')
plt.show();
```

```python
#Creamos una columna con 20 rangos de a침os.
df['a침o_bins']=pd.cut(df['year'],20)
df.head()
```

```python
categoricas=['dead','gender','specie','rangos_edad']
```

```python
#Creamos un subplot para cada variable categorica representada a lo largo de los a침os.
fig, axes = plt.subplots(4,1, figsize=(10,20))

for ax in fig.axes:
    ax.tick_params(labelrotation=45)
for i in range(len(categoricas)):
    sns.countplot(x = 'a침o_bins', data = df, ax=axes[i], hue = categoricas[i])

plt.show(); 
```

`Podemos observar que necesitamos quitar outliers para poder apreciar cambios en los ataques segun los a침os. Ya que los rangos generados en la columna 'a침os_bins' se centrarian mas en los a침os donde ocurren la mayoria de ataques (a partir del a침o 1900).`
