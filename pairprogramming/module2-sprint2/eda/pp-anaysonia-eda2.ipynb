```python
#Creamos una funcion para pasar todos los valores de la columna df['age] a valores numericos limpios.
def años (col):
    try:
        int(col)
        return int(col)       
    except:
        try:
            list= re.findall('\d+',col)
            if len(list)>1:
                suma=0
                for i in list:
                    suma+=int(i)
                    longitud=len(list)
                    return int(suma/longitud)
            elif len(list)==1:
                return int(list[0])
            else:
                return np.nan
        except:
            return np.nan
```

```python
#Aplicamos la funcion y creamos una columna nueva con los resultados de la funcion para cada celda.
df['edad']=df['age'].apply(años)
```

```python
#Observamos los valores unicos de la columna sin limpiar.
df['age'].unique()
```

```python
#Observamos los valores unicos de la columna nueva, ya limpios.
df['edad'].unique()
```
