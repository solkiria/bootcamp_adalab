- Extraed los valores únicos su frecuencia. **BONUS** presenta los resultados en un dataframe, para que quede de esta forma:

|variable   |n_uniques   |
|---|---|
|case_number   |[2018.06.25, 2018.06.18, 2018.06.09, 2018.06.0...   |
|type   |[Boating, Unprovoked, Invalid, Provoked, Quest...   |
|country   |[USA, AUSTRALIA, MEXICO, BRAZIL, ENGLAND, SOUT...   |

Donde variable es cada una de las variables categóricas que tenemos en el dataset y n_unicos es una lista con los nombres de los valores únicos de cada variable.

```python
#Creamos un for loop para obtener un diccionaio con los valores unicos y su frecuencia, para cada columna
dict={}
for col in df.columns:
    dict[col]=str(df[col].value_counts().to_dict())
```

```python
#Transformamos los datos almacenados en un dataframe para poder verlo en una tabla.
pd.DataFrame(dict, index=[0]).T
```
