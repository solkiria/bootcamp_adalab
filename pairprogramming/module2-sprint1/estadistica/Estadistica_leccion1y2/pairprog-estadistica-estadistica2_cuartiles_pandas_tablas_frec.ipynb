- Utilizando métodos de pandas: media, mediana, moda, máximo y mínimo

```python
#Escogemos dos de las columnas para la realizacion de los analisis estadisticos (media, mediana, moda, minimo y maximo).
print('La media de los valores numericos de la columna WRI es:', df['WRI'].mean())
print('La media de los valores numericos de la columna Susceptibility es:', df['Susceptibility'].mean())

print('La mediana de los valores numericos de la columna WRI es:', df['WRI'].median())
print('La mediana de los valores numericos de la columna Susceptibility es:', df['Susceptibility'].median())

print('La moda de los valores numericos de la columna WRI es:', df['WRI'].mode())
print('La moda de los valores numericos de la columna Susceptibility es:', df['Susceptibility'].mode())

print('El minimo de los valores numericos de la columna WRI es:', df['WRI'].min())
print('El minimo de los valores numericos de la columna Susceptibility es:', df['Susceptibility'].min())

print('El maximo de los valores numericos de la columna WRI es:', df['WRI'].max())
print('El maximo de los valores numericos de la columna Susceptibility es:', df['Susceptibility'].max())
```

    - En el caso de la columna WRI observamos que la media aritmetica (7.55) se encuentra proxima a la mediana (6.52), sin embargo las 2 modas de esta columna (3.46 y 3.56) de esta columna distan bastante de la media aritmetica y la mediana. Esto quiere decir que la media no representa la mayoria datos de la columna, ya que las modas son bastante inferiores a su valor.

    - En el caso de la columna Susceptibility la media aritmetica (30.73) se acerca a la mediana (25.4), pero se aleja de la moda (17.27). Observamos que, la moda que alberga los valores con mayor frecuencia, no se representada por la media aritmetica aritmetica y la mediana.

    - La media del maximo y el minimo (28.37) de la columna WRI no se corresponde con la media de esta columna (7.55).

    - La media del maximo y el minimo (39.55) de la columna Susceptibility no se corresponde con la media aunque se aproxima (30.74).

- Percentiles 25, 75

```python
#Obtenemos el percentil 25 y 75 de la siguiente manera:
percentil_25_WRI= np.percentile(df['WRI'],25)
print('El percentil 25 de la coumna WRI es:', percentil_25_WRI)

percentil_25_Susceptibility= np.percentile(df['Susceptibility'],25)
print('El percentil 25 de la coumna Susceptibility es:', percentil_25_Susceptibility)

percentil_75_WRI= np.percentile(df['WRI'],75)
print('El percentil 75 de la coumna WRI es:', percentil_75_WRI)

percentil_75_Susceptibility= np.percentile(df['Susceptibility'],75)
print('El percentil 75 de la coumna Susceptibility es:', percentil_75_Susceptibility)

```

    - Interpretamos que el 25% de valores de WRI se encuentran por debajo de 3,74 y el 75% se encuentra por debajo de 9,4.
    - Interpretamos que el 25% de valores de Susceptibility se encuentran por debajo de 17.79 y el 75% se encuentra por debajo de 42,64.

- Rango Intercualtílico

```python
#Obtenemos el rango intercuartilico para la columna WRI y Susceptibility de la siguiente manera:
q3_WRI,q1_WRI=np.percentile(df['WRI'],[75,25])
print('El rango intercuartilico de la columna WRI es:',q3_WRI - q1_WRI)
q3_Susceptibility,q1_Susceptibility=np.percentile(df['Susceptibility'],[75,25])
print('El rango intercuartilico de la columna Susceptibility es:',q3_Susceptibility - q1_Susceptibility)
```

    -   Al restarle a q1 el rango intercuartilico y posteriormente sumarle a q3 el rango intercuartilico, obtendriamos los valores a partir de los cuales identificariamos los valores extremos(outliers).

- Tabla de frecuencias
    - Variables cualitativas
    - Variables cuantitativas
    - Frecuencia acumulada
    - Representación de al menos una tabla de frecuencias

```python
#Hemos realizado la tabla de frecuencias de las variables cualitativas de la columna 'WRI Category'.
frecuencia_WRI_Category=df.groupby('WRI Category').count().reset_index()
frecuencia_WRI_Category
```

```python
#Hemos realizado la tabla de frecuencias de las variables cuantitativas de la columna 'WRI'.
frecuencia_WRI=df.groupby('WRI').count().reset_index().sort_values(by='Region', ascending=False)
frecuencia_WRI
```

```python
#Creamos una columna denominada ¨order¨ donde añadimos una lista con la numeracion que identifique el orden 
# en el que se deben representar las filas.
frecuencia_WRI_Category['order']=[3,1,2,4,0]
frecuencia_WRI_Category
```

```python
#Ordenamos el DataFrame por la columna 'order', para que las categorias 'WRI Category' esten en el orden desado (ascendente).
frecuencia_WRI_Category.sort_values('order')
```

```python
#Anadimos una columna denominada 'frec_acum_WRI_cat' para obtener la frecuencia acumulada aplicando el metodo cumsum.
frecuencia_WRI_Category['frec_acum_WRI_cat']=frecuencia_WRI_Category['WRI'].cumsum()
frecuencia_WRI_Category
```

```python
#Realizamos la representacion grafica de las frecuencias de las variables cualitativas. Nos marca los valores que corresponden a cada categoria
# en lugar de ponernos el recuento de filas en esa categoria.
plt.figure(figsize=(15,7))

fig=sns.histplot(x='WRI Category', y= 'WRI', data=frecuencia_WRI_Category)
fig.set_title('WRI Category Histplot')

plt.show()
```
