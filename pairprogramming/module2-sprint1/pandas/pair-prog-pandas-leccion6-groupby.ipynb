```python
# importamos la libreria pandas para poder utilizar sus metodos
import pandas as pd
```

```python
# creamos el dataframe a partir del archivo .csv 
df1=pd.read_csv('Data_lecc5_pandas/attacks_lec5.csv')
df1.head()
```

1. 쮺u치ntos ataques hubo por pa칤s? Devuelve un DataFrame.
游늷 NOTA Ordena los resultados de mayor a menor. Esto todav칤a no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values(). 
Aqu칤 ten칠is una explicaci칩n de c칩mo hacerlo.

```python
# en primer lugar agrupamos todos los ataques por pais.
attacks_country=df1.groupby('country')['country'].count()
attacks_country
```

```python
# ordenamos los ataques por orden descendente aplicando el metodo '.sort_values(ascending= False)'
attacks_country_orden=df1.groupby('country')['country'].count().sort_values(ascending= False)
attacks_country_orden
```

```python
# ordeamos DataFrame con los datos ordenados de mayor a menor
df_attacks_country_orden=pd.DataFrame(attacks_country_orden)
df_attacks_country_orden
```

2. 쮺u치ntos ataques ha habido por a침o? Devuelve un DataFrame. De nuevo ordena los resultados de mayor a menor.

```python
# agrupamos los datos por year y ordenamos de manera descandente
attacks_year=df1.groupby('year')['year'].count().sort_values(ascending= False)
attacks_year
```

```python
# creamos un DataFrame de la variable de la agrupacion anterior 
df_attacks_year=pd.DataFrame(attacks_year)
df_attacks_year
```

3. 쮺u치ntos ataques hubo por a침o y sexo? Devolved un DataFrame solo del a침o 2000 en adelante.

```python
# obtenemos un DataFrame con los datos de ataques posteriores al year 2000.
attacks_2000=df1[df1['year']>2000]
attacks_2000.head()
```

```python
# agrupamos los datos por year y separandolos por sexo dentro de cada year
attacks_year_sex=attacks_2000.groupby(['year','sex_'])['year'].count().sort_values(ascending=False)
df_attacks_year_sex=pd.DataFrame(attacks_year_sex)
df_attacks_year_sex
```

```python
df_USA_AUS_SA_letal_sex=USA_AUS_SA_letal_sex.reset_index()
df_USA_AUS_SA_letal_sex
```
