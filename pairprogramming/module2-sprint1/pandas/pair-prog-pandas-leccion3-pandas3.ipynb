- Sacad los valores √∫nicos y su frecuencia de las variables num√©ricas. Almacenad los resultados en un diccionario. Uno para cada variable.

```python
#obtenemos los tipos de datos de cada una de las columnas para conocer las columnas
#  con datos numericos y poder generar los diccionarios con los valores unicos y sus frecuencias.
attacks.dtypes
```

```python
#creacion de diccionario de la variable numerica 'Unnamed: 0'.
dicc_unnamed=dict(attacks['Unnamed: 0'].value_counts())
dicc_unnamed
```

```python
#creacion de diccionario de la variable numerica 'Year'.
dicc_year=dict(attacks['Year'].value_counts())
dicc_year
```

```python
#creacion de diccionario de la variable numerica 'original order'.
dicc_original_order=dict(attacks['original order'].value_counts())
dicc_original_order
```

```python
#Para automatizar el proceso el proceso anterior, se puede generar con un for loop, el
#  resultado final ser√≠a una lista de 3 diccionarios. Por lo tanto no tendr√≠amos almacenados
#  cada diccionario en una variable. Habria que sacar cada diccionario por su indice de la
#  lista y almacenarlo en una variable.
lista_dicc=[]
for x in nombres_columnas:
    if attacks[x].dtypes=='object':
        pass
    else:
        lista_dicc.append(dict(attacks[x].value_counts()))

len(lista_dicc)
```

- Cambiad el nombre de las columnas para que todas est√©n en min√∫sculas y reemplazad los espacios por _.

```python
#Creamos un diccionario empleando una list comprehension para poner en minuscula los nombres de las columnas de todo el
#  dataframe y sustituir los espacios por "_".
columnas_nuevas={columna:columna.lower().replace(' ','_') for columna in attacks.columns}
columnas_nuevas
```

```python
#Empleando el diccionario creado se aplica el metodo ".rename()" para que se haga efectivo el cambio de nombre en las columnas.
attacks.rename(columns=columnas_nuevas, inplace=True)
attacks
```

- Eliminad las columnas de `Unnamed: 22` y `Unnamed: 23`, `Case Number.1`, `Case Number.2`, `original order`, `Investigator or Source`, `pdf`, `href formula`.

```python
#Eliminamos las columnas indicadas con el metodo ".drop()".
eliminar_columnas_attacks=attacks.drop(['unnamed:_22','unnamed:_23','case_number.1','case_number.2','original_order','investigator_or_source','pdf','href_formula'],axis=1)
eliminar_columnas_attacks
```
