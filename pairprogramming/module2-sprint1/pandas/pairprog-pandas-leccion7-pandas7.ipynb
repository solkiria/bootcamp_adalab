1. Cread una columna nueva y una funci√≥n que nos de el mes en el que ocurri√≥ el ataque. Tened en cuenta que no todas las filas tienen la misma estructura y que puede que no haya la informaci√≥n de mes. En ese caso devolved un nulo (NaN).

üí° Pista üí°Pod√©is usar regex.

```python
# importamos las librerias necesarias para la utilizacion de sus metodos
import pandas as pd
import numpy as np
import re
```

```python
# creamos el dataframe a partir del archivo .csv .
df=pd.read_csv('Data_lecc5_pandas/attacks_lec5.csv')
df.head()
```

- Utilizamos los siguientesmetodos para obtener informacion sobre el DataFrame:

```python
# aplicamos el metodo '.value_counts()' a la columna date para la obtencion de los valores unicos.
df["date"].value_counts()
```

```python
# obtenemos informacion sobre los datos numericos de la columna date
df['date'].describe()
```

```python
# nos devuelve la cantidad de valores que son nulos en cada columna del DataFrame 1
df['date'].isnull().sum()
```

```python
# creamos una copia del DataFrame para trabajar sobre ella.
df2=df.copy()
```

```python
# definimos una funcion donde aplicamos un patron de regex para la busqueda del mes 
patron = "[A-Z]\w+"
def sacar_mes(col):
    try:
        return re.findall(patron, col)[0]
        
    except:
        return np.nan
```

```python
# creamos una columna nueva denominada 'month' donde almacenaremos los datos obtenidos de la funcion, aplicando el metodo '
# .apply()'en la columna date.
df2['month'] = df2['date'].apply(sacar_mes)
df2.head()
```

```python
# obtenemos los valores de la columna 'month'.
df2['month'].value_counts()
```

2. Vamos con la columna fatal_(y/n). Lo primero que ten√©is que hacer es evaluar los valores √∫nicos que hay. Esperar√≠amos tener solo dos valores: y y n, pero tristemente no ser√° as√≠ üò≠. Cread una columna nueva y una funci√≥n que devuelva √∫nicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor.

```python
# nos muestra los elementos unicos del array.
df2["fatal_(y/n)"].unique()
```

```python
# creamos la funcion para que nos devuelva los valores 'Y' o 'N', de lo contrario nos devolvera un nulo.  
def return_yn (col):
    if col == "N" or col == "Y":
        return col
    else:
        return np.nan
```

```python
# una vez hemos creado la funcion, almacenamos los datos obtenidos en dicha funcion mediante la utilizacion del metodo '.apply()' 
df2['dead']=df2["fatal_(y/n)"].apply(return_yn)
```

```python
df2['dead'].head()
```

```python
# comprobamos los elementos unicos del array tras realizar la funcion.
df2['dead'].unique()
```

3. Seguid la misma l√≥gica que en el ejercicio anterior pero con la columna sex. Pero en este caso vamos a incluir un retito m√°s: En este caso querremos tener dos valores √∫nicos F y M. De nuevo en la columna hay otros car√°cteres raros como 'M ', 'lli', 'N', '.'. En esos casos, queremos que se reemplacen por M o F de forma aleatoria (tendr√©is que usar la librer√≠a random). En caso de que sea nulo, que la funci√≥n devuelva nulo (np.nan).

```python
# importamos la libreria para la utilizacion de sus metodos
import random  
```

```python
df2["sex_"].unique()
```

```python
# creamos la funcion para que nos devuelva los valores 'F' o 'M', de lo contrario nos devolvera un nulo.  
def change_sex_ (col):
    if col == "F" or col == "M":
        return col
    else:
        return np.nan
```

```python
df2['gender']=df2['sex_'].apply(change_sex_)
df2.head()
```
