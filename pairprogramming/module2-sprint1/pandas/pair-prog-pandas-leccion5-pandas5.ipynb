2. Chequead que los nombres de las columnas ver si se llaman iguales entre dataframes . ¿Están todas en minúsculas? ¿Tienen espacios? En caso de que no sean todas iguales unificad el nombre de las columnas.

```python
# creamos un diccionario empleando una list comprehension para poner en minuscula los nombres de las columnas de todo el
#  dataframe y sustituir los espacios por "_" 
df1_columnas={col:col.lower().replace(' ','_') for col in df1.columns}
df1.rename(columns=df1_columnas,inplace=True)
df1.head()
```

```python
# creamos un diccionario empleando una list comprehension para poner en minuscula los nombres de las columnas de todo el
#  dataframe y sustituir los espacios por "_" 
df2_columnas={col:col.lower().replace(' ','_') for col in df2.columns}
df2.rename(columns=df2_columnas,inplace=True)
df2.head()
```

```python
# creamos un diccionario empleando una list comprehension para poner en minuscula los nombres de las columnas de todo el
#  dataframe y sustituir los espacios por "_" 
df3_columnas={col:col.lower().replace(' ','_') for col in df3.columns}
df3.rename(columns=df3_columnas,inplace=True)
df3.head()
```

3. En este momento nos habremos dado cuenta que el DataFrame que generamos en el pair programming Pandas IV es similar al DataFrame de los datos historicos. Si recordamos, eliminamos algunas columnas en la sesión III que no eran útiles, pero esas columnas están todavía en el histórico. Si no fuera por eso podríamos unir esos dataframes. Realizad los cambios que creais necesarios para que los dataframes sean iguales y luego juntadlos usando el mejor método.

```python
# eliminamos las columnas del DataFrame df3 con el método .drop(), que anteriormente ya habiamos eliminado en la leccion 3 en el "df1".
df3=df3.drop(['unnamed:_22','unnamed:_23','case_number.1','case_number.2','original_order','investigator_or_source','pdf','href_formula'],axis=1)
df3.head()
```

```python
# mostramos las columnas para verificar que las columnas se han eliminado correctamente en el df1
df1.columns
```

```python
# mostramos las columnas para verificar que las columnas se han eliminado correctamente en el df2
df2.columns
```

```python
# mostramos las columnas para verificar que las columnas se han eliminado correctamente en el df3
df3.columns
```

```python
#unimos los dataframes con la funcion '.concat()'
df_concat=pd.concat([df1,df3],axis=0,ignore_index=False)
df_concat.head()
```

```python
# mostramos numero de filas y columnas 
df1.shape
print('El DataFrame df1 tiene',df1.shape[0],'filas')
```

```python
df3.shape
print('El DataFrame df3 tiene',df3.shape[0],'filas')
```

```python
#podemos observar que las filas del df1 y el df3 se han unido en un solo DataFrame (df_concat1).
df_concat.shape
print('El DataFrame df_concat tiene',df_concat.shape[0],'filas')
```

4. Es el momento de unir la información de todos los ataques de tiburón que acabamos de crear con el DataFrame de información adicional. Debatid que método es el más correcto para unir toda esta información y ejecutadlo.

📌 NOTA Cuando hagáis la unión considerad que la información que más nos interesa es la del DataFrame que creamos en el apartado anterior, lo cual determinará el tipo de unión.

```python
# unimos ambos DataFrames (df_concat1 y df2) con el metodo .merge(), haciendo un left merge, ya que
#  el DataFrame que más nos interesa es el df_concat.
df_merge=pd.merge(left=df_concat,right=df2,how='left',on='case_number')
df_merge.head()
```

```python
df_concat.shape
print('El DataFrame df_concat tiene',df_concat.shape[0],'filas y',df_concat.shape[1], 'columnas.')
```

```python
df2.shape
print('El DataFrame df2 tiene',df2.shape[0],'filas y',df2.shape[1], 'columnas.')
```

```python
df_merge.shape
print('El DataFrame df_merge tiene',df_merge.shape[0],'filas y',df_merge.shape[1], 'columnas.')
#Observamos que tenemos el mismo número de filas que el df_concat (el que mas nos interesa), ya
#  que df_concat no tiene todos los registros de ataques que tenía df2. Por tanto, perdemos casos
#  de ataques (filas) pero unimos la informacion de los ataques comunes en ambos DataFrames, ya que
#  se añaden las columnas del df2 al df_concat. En este caso se han añadido 3 columnas porque la
#  columna case_number estaba presente en ambos DataFrames (es la columna de union).
```
