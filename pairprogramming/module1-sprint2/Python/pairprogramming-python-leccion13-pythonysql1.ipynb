4.- Hagamos unas queries facilitas:

Extraed el primer pedido donde se ha gastado más. Devolved el id del pedido y la cantidad gastada.

```python
conexion = mysql.connector.connect(
  host="localhost",
  user="root",
  password="AlumnaAdalab",
  database="northwind"
)

mycursor = conexion.cursor()

query = ("""SELECT order_id, (quantity*unit_price*(1-discount)) AS 'Cantidad Gastada'
            FROM order_details
            ORDER BY 'Cantidad Gastada' DESC
            LIMIT 1""")

mycursor.execute(query)

for order_id, cantidad in mycursor:
  print(f"El primer pedido {order_id} ha gastado {cantidad}")

mycursor.close()
conexion.close()
```

Devuelve la misma query que en el anterior ejercicio, pero en este caso devuelve todos los resultados.

```python
conexion = mysql.connector.connect(
  host="localhost",
  user="root",
  password="AlumnaAdalab",
  database="northwind"
)

mycursor = conexion.cursor()

# Utilizaremos Limit para que solo nos aparezcan los 10 primeros resultados ya que la query es bastante larga

query = ("""SELECT order_id, (quantity*unit_price*(1-discount)) AS 'Cantidad Gastada'
            FROM order_details
            ORDER BY 'Cantidad Gastada' DESC
            LIMIT 10""")

mycursor.execute(query)

myresult= mycursor.fetchall()
print(myresult)


mycursor.close()
conexion.close()
```

Extraed el números de pedidos que se hayan hecho por dia. Devuelve los resultados en una dataframe.

--- Para trabajar con dataframe nos tendremos que instalar e importar la librería pandas

```python
import pandas as pd
```

```python
conexion = mysql.connector.connect(
  host="localhost",
  user="root",
  password="AlumnaAdalab",
  database="northwind"
)

mycursor = conexion.cursor()

query = ("""SELECT COUNT(order_id), order_date
            FROM orders
            GROUP BY order_date
            """)

mycursor.execute(query)

myresult= mycursor.fetchall()


df = pd.DataFrame(myresult, columns = ['Pedidos', 'Fecha de Pedido'])


mycursor.close()
conexion.close()
```

Hemos utilizado el método head para que nos saque solo los primeros cinco resultados.

```python
df.head()
```

Extraer los pedidos hechos por "Nancy Daviolo". Tendréis que hacer una subquery.

```python
conexion = mysql.connector.connect(
  host="localhost",
  user="root",
  password="AlumnaAdalab",
  database="northwind"
)

mycursor = conexion.cursor()

query = ("""SELECT * FROM orders
            WHERE employee_id  IN (
						SELECT employee_id
                        FROM employees
                        WHERE first_name = 'Nancy' and last_name= 'Davolio')
                        """)

mycursor.execute(query)

myresult= mycursor.fetchall()


df = pd.DataFrame(myresult, columns = ['OrderID', 'CustomerID', 'EmployeeID','OrderDate','OrderRequired','OrderShipped', 'ShipVia', 'Freight', 'ShipName', 'ShipAddress', 'ShipCity', 'ShipRegion', 'ShipPostalCode', 'ShipCountry'])
                                      

mycursor.close()
conexion.close()
```
