```python
# Importamos las librer铆as necesarias.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
pd.options.display.max_columns=None   #Para que se muestren todas las columnas de los dataframes.
```

```python
#Creamos un dataframe con el csv de los datos de nuestro estudio.
df = pd.read_csv("files/taxis3_estandar.csv", index_col=0)
df.head(3)
```

```python
# Comenzamos con nuestro an谩lisis ANOVA
lm = ols("""tip_box ~ color + payment + pickup_borough + dropoff_borough + passengers_robust + 
   distance_robust + fare_robust + tolls_robust + total_robust""", data=df).fit()
sm.stats.anova_lm(lm)
```

```python
#Obtenemos un resumen de la recta que ajusta a neustro modelo.
lm.summary()
```

`Tras realizar la normalizaci贸n de nuestra variable respuesta (dependiente) y no conseguir obtener una muestra de datos normales, realizamos tambi茅n una estandarizaci贸n de las variables predictoras (independientes).`

`Ahora realizamos un test ANOVA para comprobar el efecto de nuestras variables predictoras (independientes) sobre nuestra variable respuesta (dependiente). En los resultados del test podemos observar que: ` 
- `el valor de R-squared es 0.873, al igual que Adj.R-squared, lo que nos indica que el 87.3 % de la variaci贸n de los datos es explicado por las variables predictoras.`
- `el coef, nos indica la tasa de cambio de la variable respuesta por cada unidad que aumenta la variable predictora. Si observamos el signo de este coeficiente podemos indicar si la relaci贸n es positiva o negativa entre la variable predictora y respuesta. Nos encontramos que tenemos una relaci贸n negativa en color[T.yellow], dropoff_borough[T.Manhattan], dropoff_borough[T.Staten Island], distance_robust, fare_robust y tolls_robust, y este signo lo que nos quiere decir es que cuando aumenta la variable predictora disminuye la variable respuesta. El resto tienen positiva, que indica, el aumento de la variable predictora hace que aumente el valor de la variable repuesta.`
- `la columna P>|t| nos indica el p-valor, que es la probabilidad de rechazar la hip贸tesis nula (H0) cuando es cierta. En este caso H0 es que la variable predictora no afecta a la variable respuesta, siendo nuestra hip贸tesis alternativa (H1) que si existe relaci贸n entre la variable predictora y la variable respuesta. Si es < 0.05 rechazamos H0 y aceptamos la hip贸tesis alternativa. Al observar los resultados del test ANOVA vemos que las variables predictoras que no afectan a la variable respuesta son 'color' y 'distance_robust' (p-valor > 0.05). En el caso de 'dropoff_borough' tenemos que una de sus categor铆as si afecta a la variable respuesta a pesar de que el resto de categor铆as no, por lo que no podemos rechazar esta variable.`
